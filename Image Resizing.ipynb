## Import Libraries

```python
import skimage
from skimage import data
from skimage import transform
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
```

### Read the image

```python
image = mpimg.imread("Imagedata/tir.jpeg")
```

```python
plt.title("original Image")
plt.imshow(image)
```

```python
image.shape
```

## Cleaning Transformations

### Reshape

```python
# preserving the height of the image and reshaping the width  and channel values 
reshaped_image = image.reshape(image.shape[0],-1)
reshaped_image.shape
```

#### 1890 :  is the width of the original image (630) multiplied by 3 number of channels 

```python
plt.figure(figsize = (12,12))
plt.title("Reshaped Image")
plt.imshow(reshaped_image)
```

### Resize#### display of the image where the 3 channels are separated 

### Resize

```python
# resize the original image to 300 by 300
image_resized = skimage.transform.resize(image,(300,300))
image_resized.shape
```

```python
plt.figure(figsize=(6,6))
plt.title("resized_image")
plt.imshow(image_resized)
```

### the resized image is distorted and this is because  during the resize we did not presereve the aspect ratio of the original image 

```python
# aspect ration =  width/height
aspect_ratio_original = image.shape[1]/float(image.shape[0])
aspect_ratio_resized = image_resized.shape[1]/float(image_resized.shape[0])
```

```python
print("Original Aspect Ratio : ", aspect_ratio_original)
print("Resized  Aspect Ratio: ", aspect_ratio_resized)
```

### Reversing color order from RGB to BGR

Used in certain frameworks such as OpenCV and Caffe2

```python
image_BGR = image[:,:,(2,1,0)]
```

```python
plt.figure(figsize=(6,6))
plt.title("BGR Image")
plt.imshow(image_BGR)
```

```python
image_BGR.shape
```

#### the shape remains the same we only fliped the order of the channels to suit certain frameworks

### Gray scale
transfroming a color image to a gray image 

```python

image_gray = skimage.color.rgb2gray(image)
```

```python
plt.imshow(image_gray, cmap = 'gray')
```
